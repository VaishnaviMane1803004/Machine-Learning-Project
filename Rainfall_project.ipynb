```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
dataset = pd.read_csv('rainfall in india 1901-2015.csv')
dataset.dropna(how='any', inplace=True)
X = dataset.iloc[:,[0,1,-1,-2,-3,-4]].values #dataset with subdivision and monthly rainfall 
Y = dataset.iloc[:,-5].values #dataset with annual rainfall
```

```python
print(X)
```

```python
Y = Y.reshape(-1,1)#to give new shape to array
print(Y)
```

```python
from sklearn.impute import SimpleImputer #scikit-learn class which is helpful in handling the missing data
imputer = SimpleImputer(missing_values=np.nan,strategy='most_frequent')#replacing null values with most frequent values
X = imputer.fit_transform(X)
Y = imputer.fit_transform(Y)
```

```python
print(X)
print(Y)
```

```python
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
X[:,0] = le1.fit_transform(X[:,0])
```

```python
print(X)
print(Y)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
print(X)
```

```python
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)#test_size - size of testing dataset
```

```python
print(X_train)
```

```python
print(Y)
```

```python
from sklearn.svm import SVR #to convert the data into linear separable form 
regressor = SVR(kernel = 'rbf')
regressor.fit(X_train, Y_train)
```

```python
regressor.score(X_train,Y_train)
```

```python
y_pred = np.array(regressor.predict(X_test),dtype=int)
Y_test = np.array(Y_test,dtype=int)
```

```python
print(y_pred)
```

```python
print(Y_test)
```

```python
y_pred = y_pred.reshape(-1,1)
Y_test = Y_test.reshape(-1,1)
```

```python
df = np.concatenate((Y_test,y_pred),axis=1)
dataframe = pd.DataFrame(df,columns=['Actual Rainfall','Predicted Rainfall'])
```

```python
print(dataframe)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,y_pred)
```
