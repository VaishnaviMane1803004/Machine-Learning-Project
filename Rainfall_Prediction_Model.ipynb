# Exploring the data

import pandas as pd
import numpy as np


```python
df = pd.read_csv("rainfall in india 1901-2015.csv")
df.head()
```

```python
df.info() #gives summary of dataframe
```

```python
df.describe() #calculates statistical data
```

```python
unique_sd = df['SUBDIVISION'].unique() #unique values of subdivision column
unique_sd
```

# Seasonal variation across India

```python
from matplotlib import pyplot as plt
drop_columns = ['ANNUAL','Jan-Feb','Mar-May','Jun-Sep','Oct-Dec']
fig, ax = plt.subplots()
(df.groupby(by='YEAR').sum().drop(drop_columns, axis=1).T.plot(alpha=0.1, figsize=(12, 6), legend=False, fontsize=12, ax=ax))
ax.set_xlabel('Months', fontsize=10)
ax.set_ylabel('Rainfall (in mm)', fontsize=10)
```

```python
import seaborn as sns
plt.figure(figsize=(10,5))
sns.heatmap(df[['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC','ANNUAL']].corr(),annot=True)
plt.show()
```

```python
corr= df.corr()
sns.clustermap(corr)
```

# Annual Prediction Rainfall

```python
df.groupby('YEAR').sum()[['ANNUAL']].plot(figsize=(10,10))
```

# Monthly Variation Of Rainfall

```python
df[['YEAR', 'JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']].groupby("YEAR").sum().plot(figsize=(13,8))
```
