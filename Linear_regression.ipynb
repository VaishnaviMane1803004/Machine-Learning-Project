# Exploratory Data Analysis

```python
dataset = pd.read_csv('rainfall in india 1901-2015.csv')
```

```python
dataset.dropna(how='any', inplace=True)
```

```python
X = dataset.iloc[:,[0,1,-1,-2,-3,-4]].values
Y = dataset.iloc[:,-5].values
```

```python
print(X)
```

```python
Y = Y.reshape(-1,1)
```

```python
print(Y)
```

# Dealing with invalid dataset

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan,strategy='most_frequent')
X = imputer.fit_transform(X)
Y = imputer.fit_transform(Y)
```

```python
print(X)
```

```python
print(Y)
```

# Encoding dataset

```python
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
X[:,0] = le1.fit_transform(X[:,0])
```

```python
print(X)
```

```python

print(Y)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
print(X)
```

```python
print(Y)
```

# Splitting Dataset into Training and Testing data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
print(X_train)
```

```python
print(Y_train)
```

# Training the model

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
regressor.score(X_train,Y_train)
```

```python
y_pred = np.array(regressor.predict(X_test),dtype=int)
Y_test = np.array(Y_test,dtype=int)
```

```python
print(y_pred)
```

```python
print(Y_test)
```

```python
y_pred = y_pred.reshape(-1,1)
Y_test = Y_test.reshape(-1,1)
```

```python
df = np.concatenate((Y_test,y_pred),axis=1)
dataframe = pd.DataFrame(df,columns=['Actual Rainfall','Predicted Rainfall'])
```

```python
print(dataframe)
```

# Calculating Accuracy

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,y_pred)
```
